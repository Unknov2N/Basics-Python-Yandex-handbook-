```python
import pandas as pd


def length_stats(text):
    symbols_to_sub = '0123456789,.;:-!@#$%^&*()_+='
    text_list = sorted(list(set(''.join([c for c in text.lower() if c not in symbols_to_sub]).split())))
    list_odd = [item for item in text_list if len(item) % 2]
    list_even = [item for item in text_list if not len(item) % 2]
    return pd.Series([len(item) for item in list_odd], index=list_odd, dtype='int64'), \
        pd.Series([len(item) for item in list_even], index=list_even, dtype='int64')
```

```python
import pandas as pd


def cheque(price_list, **products):
    columns = ['product', 'price', 'number']
    df = pd.DataFrame(columns=columns)

    for item in products:
        new_row = {'product': item, 'price': price_list[item], 'number': products[item]}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    df['cost'] = df['price'] * df['number']
    return df.sort_values(['product']).reset_index(drop=True)


products = ['bread', 'milk', 'soda', 'cream']
prices = [37, 58, 99, 72]
price_list = pd.Series(prices, products)
result = cheque(price_list, soda=3, milk=2, cream=1)
print(result)
```

```python
import pandas as pd


def cheque(price_list, **products):
    columns = ['product', 'price', 'number']
    df = pd.DataFrame(columns=columns)

    for item in products:
        new_row = {'product': item, 'price': price_list[item], 'number': products[item]}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    df['cost'] = df['price'] * df['number']
    return df.sort_values(['product']).reset_index(drop=True)


def discount(cheque: pd.DataFrame):
    df = pd.DataFrame.copy(cheque)
    df.loc[df['number'] > 2, 'cost'] /= 2
    return df


products = ['bread', 'milk', 'soda', 'cream']
prices = [37, 58, 99, 72]
price_list = pd.Series(prices, products)
result = cheque(price_list, soda=3, milk=2, cream=1)
with_discount = discount(result)
print(result)
print(with_discount)
```

#### J. Экстремум функции
Экстремум в математике — максимальное или минимальное значение функции на заданном множестве.

Чаще всего математики для поиска экстремума функции прибегают к её дифференцированию. Однако мы можем обойти этот трудоёмкий процесс и схитрить.

Напишите три функции:

values(func, start, end, step), строящую Series значений функции в точках диапазона и принимающую:

max_extremum(data) возвращает точку, в который был достигнут максимум на диапазоне.

```python
import numpy as np
import pandas as pd


def values(func, start, end, step):
    index = np.arange(start, end + step, step)
    data = np.array([func(i) for i in index])
    return pd.Series(data, index=index)


def min_extremum(data):
    dot = np.array([data.index[0], data.at[data.index[0]]])
    for i in data.index:
        # print(i, type(i), data.at[i], dot)
        if data.at[i] < dot[1]:
            dot[1] = data.at[i]
            dot[0] = i
    return dot[0]


def max_extremum(data):
    dot = np.array([data.index[0], data.at[data.index[0]]])
    for i in data.index:
        if data.at[i] > dot[1]:
            dot[1] = data.at[i]
            dot[0] = i
    return dot[0]


data = values(lambda x: x ** 2 + 2 * x + 1, -1.5, 1.7, 0.1)
print(data)
print(min_extremum(data))
print(max_extremum(data))
```
